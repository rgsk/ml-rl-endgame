```python
from typing import Any
import torch

# super simple little MLP
net: Any = torch.nn.Sequential(
    torch.nn.Linear(16, 32), torch.nn.GELU(), torch.nn.Linear(32, 1)
)
torch.random.manual_seed(42)
x = torch.randn(4, 16)
y = torch.randn(4, 1)
net.zero_grad()
yhat = net(x)
loss = torch.nn.functional.mse_loss(yhat, y)
loss.backward()
print(net[0].weight.grad.view(-1)[:10])

# the loss objective here is (due to reduction='mean')
# L = 1/4 * [
#               (y[0] - yhat[0])**2 +
#               (y[1] - yhat[1])**2 +
#               (y[2] - yhat[2])**2 +
#               (y[3] - yhat[3])**2
#           ]
#
# NOTE: 1/4!
```

```python
net.zero_grad()
for i in range(4):
    yhat = net(x[i])
    loss = torch.nn.functional.mse_loss(yhat, y[i])
    loss = loss / 4
    loss.backward()
print(net[0].weight.grad.view(-1)[:10])
```
